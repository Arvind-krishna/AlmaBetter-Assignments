## 1. Write a Python program to create a list whose ith element is the maximum of the first i elements from a input list.

```python
# Write Your code here.
def maxfn(arr):
  maxim=min(arr)
  for i,elem in enumerate(arr):
    if elem>maxim:
      maxim=elem
    arr[i]=maxim
  return arr
    


#arr=[ 0 , -1 , 3 , 8 , 5 , 9 , 8 , 14 , 2 , 4 , 3 , -10 , 10 , 17 , 41 , 22 , -4 , -4 , -15 , 0 ]
arr=[6,5,4,3,2,1]

print(maxfn(arr))
```

## 4. From the above dataframes, write a SQL query to find those retailers who have bought 'key board' but not 'mouse'. 
## Return retailer ID.

**Expected Output:** 

image.png

```python
# Write Your code here.
pd.read_sql_query ('''SELECT retailer_id
FROM sales_info 
JOIN item 
USING(item_code)
WHERE item_desc IN ('mouse', 'key board')
GROUP BY retailer_id
HAVING SUM(DISTINCT item_code) = 
(SELECT item_code FROM item WHERE item_desc = 'key board');''',conn)
```

## 7. What is the curse of dimensionality? Can you list some ways to deal with it?

Post the data preparation phase, we provide the training data for our ML model to train on. 

However, if there are too many dimensions/features in the training data, it would increase the complexity of the learning process significantly, often leading to situations where it may not even be feasible to go with. This is called the curse of dimensionality.



To deal with this, we must put efforts towards reducing the dimensionality of the data. 

A basic approach, is to observe the collinearity between features, and disregard one feature from a pair/s of features that are highly collinear.

Another popular approach, is to use principal component analysis (PCA) which is a technique that helps in reducing the dimensionality of the data.

The intuitive difference between L1 and L2 is that L2 helps us reduce only the overfitting of the model while keeping all the features present in the model.

It reduces the complexity of the model by shrinking the coefficients for the least important predictors very close to zero. 

L1 regression helps in reducing the problem of overfitting in the model by removing the unnecessary features, it forces the coefficient estimates to be exactly equal to zero.

As to which one one is preferred, the pros-to-cons ratio is in favor of L1.
